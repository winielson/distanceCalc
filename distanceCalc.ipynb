### Find minimum distance (km) from tornado line to location point and then store values in location df

### Formula: https://stackoverflow.com/questions/20231258/minimum-distance-between-a-point-and-a-line-in-latitude-longitude

```python
for i, loc in locDF.iterrows():
    
    for j, tor in torDF.iterrows():

        # Calculate Bearings
        y = math.sin(tor.p_elon - loc.Longitude) * math.cos(tor.p_elat)
        x = math.cos(loc.Latitude) * math.sin(tor.p_elat) - math.sin(loc.Latitude) * math.cos(tor.p_elat) * math.cos(tor.p_elat - loc.Latitude)
        bearing1 = math.degrees(math.atan2(y, x))
        bearing1 = 360 - ((bearing1 + 360) % 360)

        y2 = math.sin(tor.slon - loc.Longitude) * math.cos(tor.slat)
        x2 = math.cos(loc.Latitude) * math.sin(tor.slat) - math.sin(loc.Latitude) * math.cos(tor.slat) * math.cos(tor.slat - loc.Latitude)
        bearing2 = math.degrees(math.atan2(y2, x2))
        bearing2 = 360 - ((bearing2 + 360) % 360)

        # Find location latitude to predicted latitude distance using spherical law of cosines
        lat1Rads = math.radians(loc.Latitude)
        lat3Rads = math.radians(tor.p_elat)
        dLon = math.radians(tor.p_elon - loc.Longitude)

        # Find cross-track distance (6371 == Earth Radius)
        distanceAC = math.acos(math.sin(lat1Rads) * math.sin(lat3Rads)+math.cos(lat1Rads)*math.cos(lat3Rads)*math.cos(dLon)) * 6371
        min_distance = math.fabs(math.asin(math.sin(distanceAC/6371)*math.sin(math.radians(bearing1)-math.radians(bearing2))) * 6371)
        
        #print('min_distance: ', min_distance)
        
        # Set minimum distance variable (km)
        if j == 0:
            locDF.at[i, 'minDist'] = min_distance
            locDF.at[i, 'Index'] = tor['index']# Save index of row with min_dist to loc df
        
        if min_distance < locDF.at[i, 'minDist']:
            locDF.at[i, 'minDist'] = min_distance
            locDF.at[i, 'Index'] = tor['index']# Save index of row with min_dist to loc df
```
